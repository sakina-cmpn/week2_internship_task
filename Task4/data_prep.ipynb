```python
# Select numeric columns only
df_numeric = df.select_dtypes(include='number')

# Reshape for seaborn
df_melted = df_numeric.melt(var_name='Columns', value_name='Values')

# Plot the boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='Columns', y='Values', data=df_melted)
plt.title("Boxplot for All Numeric Columns in all_season_summary.csv")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(df['season'])
plt.show()
```

```python
# Step 1: Calculate IQR for venue_id
Q1 = df['venue_id'].quantile(0.25)
Q3 = df['venue_id'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Filter the DataFrame
df_no_outliers = df[(df['venue_id'] >= lower_bound) & (df['venue_id'] <= upper_bound)]

# Step 4 (optional): Save the cleaned file
df_no_outliers.to_csv('cleaned_no_outliers.csv', index=False)

print(f"Original rows: {len(df)}, After removing outliers: {len(df_no_outliers)}")

sns.boxplot(df_no_outliers['venue_id'])
plt.show()
```
